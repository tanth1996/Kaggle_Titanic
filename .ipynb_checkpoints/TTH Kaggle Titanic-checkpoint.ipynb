```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pass_df = pd.read_csv('train.csv') # passenger dataframe
pass_df.head()
```

```python
pass_df.info()
```

Age and Cabin are missing values

```python
# Ensure Passenger IDs are not duplicated
pass_df['PassengerId'].is_unique
```

```python
# Set new index as Passenger ID
pass_df.set_index('PassengerId', inplace=True)
pass_df
```

```python
# Replace 'Sex' column with one-hot variables
pass_df = pass_df.join(pd.get_dummies(pass_df['Sex']))
pass_df.drop(columns='Sex', inplace=True)
pass_df
```

```python
# Get numeric columns
pass_df_num = pass_df.select_dtypes(include=np.number)
pass_df_num
```

```python
# Standardise appropriate columns
from sklearn.preprocessing import StandardScaler
normaliser = StandardScaler(copy=True, with_mean=True, with_std=True)
std_features = ['Age','Fare']
pass_df_num_n = pass_df_num.copy()
pass_df_num_n.loc[:,std_features] = normaliser.fit_transform(pass_df_num.loc[:,std_features])

# Drop label and fill missing age values
pass_df_num_n.drop(columns='Survived', inplace=True)
pass_df_num_n.fillna(value={'Age':0})
pass_df_num_n
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)

# Fit on numeric data with missing age values replaced with 0 mean
pass_df_num_n_0 = pass_df_num_n.fillna(value={'Age':0})
pca.fit(pass_df_num_n_0)
print('Explained variances: ',pca.explained_variance_)
print('Components: ',pca.components_)
```

```python
range(1,len(pca.explained_variance_ratio_)+1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
xticks = range(1,len(pca.explained_variance_ratio_) + 1)
ax.plot(xticks, np.cumsum(pca.explained_variance_ratio_))
ax.set_xticks(xticks)
ax.grid()
ax.set_title('Cumulative explained variance of Principal Components')
ax.set_xlabel('No. of Principal Components')
plt.show()
```

```python
# Transform data into Principal Components
pass_df_pc=pd.DataFrame(pca.transform(pass_df_num_n_0))
pass_df_pc.set_index(pass_df.index, inplace=True)
pass_df_pc.rename(columns=lambda x: 'PC' + str(int(x)+1), inplace=True)
pass_df_pc
```

```python
# Append 'Survived' column for plotting
pass_df_pc = pass_df.join(pass_df_pc)
pass_df_pc
```

```python
# Plot PC components
import matplotlib.cm as cm
cmap = cm.viridis(np.linspace(0,1,2))
fig, ax = plt.subplots(1,1, figsize=(15,10))

ax.scatter(pass_df_pc['PC1'], pass_df_pc['PC2'], c=cmap[pass_df_pc['Survived']])
```
