```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import *
import graphviz
```

```python
pass_df = pd.read_csv('train.csv') # passenger dataframe
pass_df.head(20)
```

```python
pass_df.info()
```

Age and Cabin are missing values

```python
# Ensure Passenger IDs are not duplicated
pass_df['PassengerId'].is_unique
```

```python
# Set new index as Passenger ID
pass_df.set_index('PassengerId', inplace=True)
pass_df
```

```python
# Replace 'Sex' column with one-hot variables
pass_df = pass_df.join(pd.get_dummies(pass_df['Sex']))
pass_df.drop(columns='Sex', inplace=True)
pass_df
```

```python
# Drop irrelevant label
pass_df.rename(columns={'female':'IsFemale'}, inplace=True)
pass_df.drop(columns=['male'], inplace=True)
```

```python
# Get numeric columns
pass_df_num = pass_df.select_dtypes(include=np.number)
pass_df_num
```

```python
# Plot distributions of numeric variables
for column in pass_df_num:
    fig, ax = plt.subplots(1,1,figsize=(10,5))
    hist, bins, patches = ax.hist(pass_df_num[column], align='mid', rwidth=0.9)
    ax.set_xticks(bins)
    ax.set_title('Histogram of ' + column)
    ax.grid()
    plt.show()
```

```python
# Standardise appropriate columns
from sklearn.preprocessing import StandardScaler
normaliser = StandardScaler(copy=True, with_mean=True, with_std=True)
std_features = ['Age','Fare']
pass_df_num_n = pass_df_num.copy()
pass_df_num_n.loc[:,std_features] = normaliser.fit_transform(pass_df_num.loc[:,std_features])

# Fill missing age values
pass_df_num_n.fillna(value={'Age':pass_df_num_n['Age'].mean()})
pass_df_num_n
```

```python
# Fill missing age values with mean
pass_df_num_n0 = pass_df_num_n.fillna(value={'Age':pass_df_num_n['Age'].mean()})
pass_df_num_n0.info()
```

# Check correlation matrix
feature_corr = pass_df_num_n0.corr()
fig, ax = plt.subplots(1,1,figsize=(10,10))
sns.heatmap(feature_corr, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()

# Peek most correlated variables
feature_corr['Survived'].abs().sort_values(ascending=False)

```python
# Perform PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)

# Fit on numeric data
pca.fit(pass_df_num_n0)
print('Explained variances: ', pca.explained_variance_ratio_)
print('Components: ', pca.components_)

# Cumulative explained variance
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
xticks = range(1,len(pca.explained_variance_ratio_) + 1)
ax.plot(xticks, np.cumsum(pca.explained_variance_ratio_))
ax.set_xticks(xticks)
ax.grid()
ax.set_title('Cumulative explained variance of Principal Components')
ax.set_xlabel('No. of Principal Components')
plt.show()
```

```python
# Transform data into Principal Components
pass_df_pc=pd.DataFrame(pca.transform(pass_df_num_n0))
pass_df_pc.set_index(pass_df.index, inplace=True)
pass_df_pc.rename(columns=lambda x: 'PC' + str(int(x)+1), inplace=True)
pass_df_pc
```

```python
# Append 'Survived' column for plotting
pass_df_pc = pass_df_num_n0.join(pass_df_pc)
pass_df_pc = pass_df_pc.join(pass_df['Survived'])
pass_df_pc
```

```python
# Plot PC components
import matplotlib.cm as cm
cmap = cm.viridis(np.linspace(0,1,2))
fig, ax = plt.subplots(1,1, figsize=(15,10))

ax.scatter(pass_df_pc['PC1'], pass_df_pc['PC2'], c=cmap[pass_df_pc['Survived']])
plt.show()
```

### No significant separation of survivors and fatalities based on PC1 and PC2

```python
# Split data into training and validation
from sklearn.model_selection import train_test_split
Features = pass_df_pc.drop(columns=['Survived'])
Label = pass_df_pc[['Survived']]
TrainFeat, TestFeat, TrainLabel, TestLabel = train_test_split(Features, Label, test_size=0.2, random_state=42)
```

```python
TrainFeat.select_dtypes(include=np.number)
```

### Out of all the Principal Components, PC5 and PC3 are the most correlated with survival
#### Let's plot them out

```python
# Plot PC components
import matplotlib.cm as cm
cmap = cm.viridis(np.linspace(0,1,2))
fig, ax = plt.subplots(1,1, figsize=(15,10))

ax.scatter(pass_df_pc['PC2'], pass_df_pc['PC3'], c=cmap[pass_df_pc['Survived']])
plt.show()
```

```python
TrainFeat.info()
```

```python
# Try a decision tree classifier
import sklearn
dtree = sklearn.tree.DecisionTreeClassifier(max_depth=None, min_samples_leaf=20)
features = ['IsFemale', 'Pclass', 'Fare', 'Parch']#,'Age','SibSp']
dtree.fit(TrainFeat.loc[:,features], TrainLabel)
dot_data = sklearn.tree.export_graphviz(dtree, out_file=None, 
                      feature_names=features,  
                      class_names=['Fatality','Survivor'],  
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
# Check decision tree accuracy on training set
TestPredict = dtree.predict(TrainFeat.loc[:,features])
TestCompare = TestPredict == TrainLabel['Survived']
Accuracy =  TestCompare.sum()/TestCompare.size
print('Decision tree accuracy on training: ', Accuracy)

# Check decision tree accuracy on test set
TestPredict = dtree.predict(TestFeat.loc[:,features])
TestCompare = TestPredict == TestLabel['Survived']
Accuracy =  TestCompare.sum()/TestCompare.size
print('Decision tree accuracy on test: ', Accuracy)

# Plot confusion matrix
Confusion = sklearn.metrics.confusion_matrix(TestLabel, TestPredict)
fig, ax = plt.subplots(1,2,figsize=(10,5))
sns.heatmap(Confusion, annot=True, ax=ax[0])
bottom, top = ax[0].get_ylim()
ax[0].set_ylim(bottom + 0.5, top - 0.5)
# Normalised confusion matrix
Confusion_n = Confusion/Confusion.sum(axis=1)[:,np.newaxis]
sns.heatmap(Confusion_n, annot=True, ax=ax[1])
bottom, top = ax[1].get_ylim()
ax[1].set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
# Try decision tree with Principal Components
dtree_PCA = sklearn.tree.DecisionTreeClassifier(max_depth=None, min_samples_leaf=15)
features_PCA = ['IsFemale', 'Pclass', 'Fare', 'Parch', 'PC2', 'PC3','PC5','PC1']
dtree_PCA.fit(TrainFeat.loc[:,features_PCA], TrainLabel)
dot_data = sklearn.tree.export_graphviz(dtree_PCA, out_file=None, 
                      feature_names=features_PCA,  
                      class_names=['Fatality','Survivor'],  
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
# Check decision tree accuracy on training set
TestPredict = dtree_PCA.predict(TrainFeat.loc[:,features_PCA])
TestCompare = TestPredict == TrainLabel['Survived']
Accuracy =  TestCompare.sum()/TestCompare.size
print('Decision tree accuracy on training: ', Accuracy)

# Check decision tree accuracy on test set
TestPredictPCA = dtree_PCA.predict(TestFeat.loc[:,features_PCA])
TestComparePCA = TestPredictPCA == TestLabel['Survived']
Accuracy =  TestComparePCA.sum()/TestComparePCA.size
print('Decision tree accuracy on test with PCs: ', Accuracy)

# Plot confusion matrix
Confusion = sklearn.metrics.confusion_matrix(TestLabel, TestPredictPCA)
fig, ax = plt.subplots(1,2,figsize=(10,5))
sns.heatmap(Confusion, annot=True, ax=ax[0])
bottom, top = ax[0].get_ylim()
ax[0].set_ylim(bottom + 0.5, top - 0.5)
# Normalised confusion matrix
Confusion_n = Confusion/Confusion.sum(axis=1)[:,np.newaxis]
sns.heatmap(Confusion_n, annot=True, ax=ax[1])
bottom, top = ax[1].get_ylim()
ax[1].set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
